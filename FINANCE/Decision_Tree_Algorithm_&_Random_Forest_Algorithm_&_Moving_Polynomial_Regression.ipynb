```python
"""
This program is written by Salar Delavar Ghashaghaei (Qashqai)
"""
import numpy as np
# Generate random data
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.rand(80) * 0.5
```

```python
"""
Decision Tree Algorithm
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Fit Decision Tree Regressor
decision_tree = DecisionTreeRegressor(max_depth=5)
decision_tree.fit(X, y)

# Predict using Decision Tree
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_pred = decision_tree.predict(X_test)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X_test, y_pred, color='cornflowerblue', label='Prediction')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Decision Tree Regression')
plt.legend()
plt.show()
```

```python
"""
Random Forest Algorithm
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Fit Random Forest Regressor
random_forest = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
random_forest.fit(X, y)

# Predict using Random Forest
y_pred_rf = random_forest.predict(X_test)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X_test, y_pred_rf, color='cornflowerblue', label='Prediction')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Random Forest Regression')
plt.legend()
plt.show()
```

```python
"""
Moving Polynomial Regression
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Moving Polynomial Regression function
def moving_polynomial_regression(X, y, period, degree):
    y_pred = np.zeros_like(y)
    poly = PolynomialFeatures(degree)
    for i in range(len(X)):
        start = max(0, i - period)
        end = min(len(X), i + period)
        X_subset = X[start:end]
        y_subset = y[start:end]
        X_poly = poly.fit_transform(X_subset)
        model = LinearRegression().fit(X_poly, y_subset)
        y_pred[i] = model.predict(poly.transform(X[i:i+1]))
    return y_pred

# Predict using Moving Polynomial Regression
period = 10
degree = 2
y_pred_poly = moving_polynomial_regression(X, y, period, degree)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X, y_pred_poly, color='cornflowerblue', label='Moving Polynomial Regression')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Moving Polynomial Regression')
plt.legend()
plt.show()
```
