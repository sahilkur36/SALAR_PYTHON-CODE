```python

import numpy as np
# Generate random data
X = np.sort(5 * np.random.rand(120, 1), axis=0)
y = np.sin(X).ravel() + np.random.rand(120) * 0.5
```

```python
"""
Decision Tree Algorithm
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Fit Decision Tree Regressor
decision_tree = DecisionTreeRegressor(max_depth=5)
decision_tree.fit(X, y)

# Predict using Decision Tree
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_pred = decision_tree.predict(X_test)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X_test, y_pred, color='cornflowerblue', label='Prediction')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Decision Tree Regression')
plt.legend()
plt.show()
```

```python
"""
Random Forest Algorithm
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Fit Random Forest Regressor
random_forest = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
random_forest.fit(X, y)

# Predict using Random Forest
y_pred_rf = random_forest.predict(X_test)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X_test, y_pred_rf, color='cornflowerblue', label='Prediction')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Random Forest Regression')
plt.legend()
plt.show()
```

```python
"""
Moving Polynomial Regression
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Moving Polynomial Regression function
def moving_polynomial_regression(X, y, period, degree):
    y_pred = np.zeros_like(y)
    poly = PolynomialFeatures(degree)
    for i in range(len(X)):
        start = max(0, i - period)
        end = min(len(X), i + period)
        X_subset = X[start:end]
        y_subset = y[start:end]
        X_poly = poly.fit_transform(X_subset)
        model = LinearRegression().fit(X_poly, y_subset)
        y_pred[i] = model.predict(poly.transform(X[i:i+1]))
    return y_pred

# Predict using Moving Polynomial Regression
period = 10
degree = 2
y_pred_poly = moving_polynomial_regression(X, y, period, degree)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X, y_pred_poly, color='cornflowerblue', label='Moving Polynomial Regression')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Moving Polynomial Regression')
plt.legend()
plt.show()
```

```python
"""
Exponential Regression
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Exponential Regression function
def exponential_regression(X, y):
    model = LinearRegression()
    X_transformed = X
    y_transformed = np.log(y + 1)  # Transform y by taking the log
    model.fit(X_transformed, y_transformed)
    y_pred = np.exp(model.predict(X_transformed)) - 1  # Reverse the transformation
    return y_pred

# Predict using Exponential Regression
y_pred_exp = exponential_regression(X, y)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X, y_pred_exp, color='cornflowerblue', label='Exponential Regression')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Exponential Regression')
plt.legend()
plt.show()
```

```python
"""
Logarithmic Regression
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate random data
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.rand(80) * 0.5

# Logarithmic Regression function
def logarithmic_regression(X, y):
    model = LinearRegression()
    X_transformed = np.log(X + 1)  # Transform X by taking the log
    y_transformed = y
    model.fit(X_transformed, y_transformed)
    y_pred = model.predict(X_transformed)
    return y_pred

# Predict using Logarithmic Regression
y_pred_log = logarithmic_regression(X, y)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X, y_pred_log, color='cornflowerblue', label='Logarithmic Regression')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Logarithmic Regression')
plt.legend()
plt.show()
```

```python
"""
Logistic Regression
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Convert target to binary for logistic regression
y_binary = (y > 0.5).astype(int)

# Logistic Regression function
def logistic_regression(X, y):
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    model = LogisticRegression()
    model.fit(X_scaled, y)
    y_pred = model.predict_proba(X_scaled)[:, 1]
    return y_pred

# Predict using Logistic Regression
y_pred_logistic = logistic_regression(X, y_binary)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y_binary, color='darkorange', label='Data')
plt.plot(X, y_pred_logistic, color='cornflowerblue', label='Logistic Regression')
plt.xlabel('Data')
plt.ylabel('Probability')
plt.title('Logistic Regression')
plt.legend()
plt.show()
```

```python
"""
Ridge Regression
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

# Ridge Regression function
def ridge_regression(X, y, alpha=1.0):
    model = Ridge(alpha=alpha)
    model.fit(X, y)
    y_pred = model.predict(X)
    return y_pred

# Predict using Ridge Regression
y_pred_ridge = ridge_regression(X, y)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X, y_pred_ridge, color='cornflowerblue', label='Ridge Regression')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Ridge Regression')
plt.legend()
plt.show()
```

```python
"""
Lasso Regression
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso

# Lasso Regression function
def lasso_regression(X, y, alpha=0.1):
    model = Lasso(alpha=alpha)
    model.fit(X, y)
    y_pred = model.predict(X)
    return y_pred

# Predict using Lasso Regression
y_pred_lasso = lasso_regression(X, y)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X, y_pred_lasso, color='cornflowerblue', label='Lasso Regression')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Lasso Regression')
plt.legend()
plt.show()
```

```python
"""
Quantile Regression
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import QuantileRegressor

# Quantile Regression function
def quantile_regression(X, y, quantile=0.5):
    model = QuantileRegressor(quantile=quantile)
    model.fit(X, y)
    y_pred = model.predict(X)
    return y_pred

# Predict using Quantile Regression
y_pred_quantile = quantile_regression(X, y)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X, y_pred_quantile, color='cornflowerblue', label='Quantile Regression')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Quantile Regression')
plt.legend()
plt.show()
```

```python
"""
Moving Quantile Regression
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import QuantileRegressor

# Moving Quantile Regression function
def moving_quantile_regression(X, y, period, quantile=0.5):
    y_pred = np.zeros_like(y)
    for i in range(len(X)):
        start = max(0, i - period)
        end = min(len(X), i + period)
        X_subset = X[start:end]
        y_subset = y[start:end]
        model = QuantileRegressor(quantile=quantile, alpha=0).fit(X_subset, y_subset)
        y_pred[i] = model.predict(X[i:i+1])
    return y_pred

# Predict using Moving Quantile Regression
period = 10
quantile = 0.5  # Median quantile
y_pred_quantile = moving_quantile_regression(X, y, period, quantile)


# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X, y_pred_quantile, color='cornflowerblue', label='Moving Quantile Regression')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Moving Quantile Regression')
plt.legend()
plt.show()
```

```python
"""
Bayesian Regression
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import BayesianRidge

# Bayesian Regression function
def bayesian_regression(X, y):
    model = BayesianRidge()
    model.fit(X, y)
    y_pred = model.predict(X)
    return y_pred

# Predict using Bayesian Regression
y_pred_bayesian = bayesian_regression(X, y)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X, y_pred_bayesian, color='cornflowerblue', label='Bayesian Regression')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Bayesian Regression')
plt.legend()
plt.show()
```
